# Problem 1 Optimisation

```python
import numpy as np
from numpy import linalg as LA
```

As we refview on class with Dr Sebastien, the $\mu$ vector is:

```python
mu = 0.01*np.array([2, 7, 15, 20])
mu= mu.T
print(mu)
```

On the other hand, $\sigma$ is:

```python
sigma = 0.01*np.array([5, 12, 17, 25])
sigma=sigma.T
print(sigma)
```

Finally, the correlation matrix

```python
R= np.array([[1,0.3, 0.3,0.3],[0.3,1,0.6,0.6],[0.3,0.6,1,0.6],[0.3,0.6,0.6,1]])
print(R)
```

```python
Sigma_n = np.outer(sigma, sigma) * R
print(Sigma_n)
```

Now is time to get the intermediate values "A", "B" and "C"

```python
n= len(mu)

ones_n = np.ones(n)

A= ones_n @ LA.inv(Sigma_n) @ ones_n.T
B= mu @ LA.inv(Sigma_n) @ ones_n.T
C= mu @ LA.inv(Sigma_n) @ mu.T

print(A)
print(B)
print(C)
```

Now, the lagrange multipliers:

```python
m= 0.045 # The return expected in out portffolio

Denom= A*C-B**2

lambda_Lagr = (A*m-B)/ Denom

gamma_Lagr = (C-B*m)/ Denom

print(lambda_Lagr)
print(gamma_Lagr)
```

Lastly, the optimat asset allocation w:

```python
weights =  LA.inv(Sigma_n) @ (lambda_Lagr*mu+gamma_Lagr*ones_n)

print(weights)
```

Verifying the constraints:

```python

print(round(mu@weights.T,4))

print(round(sum(weights),4))

```
