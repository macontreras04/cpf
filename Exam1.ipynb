# Solution
As we refview on class with Dr Sebastien, the $\mu$ vector is:

```python
mu = 0.01*np.array([2, 7, 15, 20])
mu= mu.T
print(mu)
```

On the other hand, $\sigma$ is:

```python
sigma = 0.01*np.array([5, 12, 17, 25])
sigma=sigma.T
print(sigma)
```

Finally, the correlation matrix

```python
R= np.array([[1,0.3, 0.3,0.3],[0.3,1,0.6,0.6],[0.3,0.6,1,0.6],[0.3,0.6,0.6,1]])
print(R)
```

```python
Sigma_n = np.outer(sigma, sigma) * R
print(Sigma_n)
```

Now is time to get the intermediate values "A", "B" and "C"

```python
n= len(mu)

ones_n = np.ones(n)

A= ones_n @ LA.inv(Sigma_n) @ ones_n.T
B= mu @ LA.inv(Sigma_n) @ ones_n.T
C= mu @ LA.inv(Sigma_n) @ mu.T

print(A)
print(B)
print(C)
```

Now, the lagrange multipliers:

```python
m= 0.045 # The return expected in out portffolio

Denom= A*C-B**2

lambda_Lagr = (A*m-B)/ Denom

gamma_Lagr = (C-B*m)/ Denom

print(lambda_Lagr)
print(gamma_Lagr)
```

Lastly, the optimat asset allocation w:

```python
weights =  LA.inv(Sigma_n) @ (lambda_Lagr*mu+gamma_Lagr*ones_n)

print(weights)
```

Verifying the constraints:

```python

print(round(mu@weights.T,4))

print(round(sum(weights),4))

```

## Problem 2: VaR and ES

```python
import numpy as np
from scipy.stats import norm
```

```python
mu = np.zeros(3)
mu= mu.T
print(mu)
```

```python
sigma = 0.01*np.array([30, 20, 15])
sigma=sigma.T
print(sigma)
```

```python
R= np.array([[1,0.8, 0.5],[0.8,1,0.3],[0.5,0.3,1]])
print(R)
```

```python
Sigma_n = np.outer(sigma, sigma) * R
print(Sigma_n)
```

```python
w = 0.01*np.array([50, 20, 30])
w = w.T
print(w)
```

Here the equations will be displayed

```python
a= 0.99
z_a = -norm.ppf(a)
print(z_a)
```

```python
import math as mt

VaRsens_w1= round(mu[0] + z_a * (Sigma_n@w)[0] * (1/mt.sqrt(w @ Sigma_n @ w.T)),4)
VaRsens_w2= round(mu[1] + z_a * (Sigma_n@w)[1] * (1/mt.sqrt(w @ Sigma_n @ w.T)),4)
VaRsens_w3= round(mu[2] + z_a * (Sigma_n@w)[2] * (1/mt.sqrt(w @ Sigma_n @ w.T)),4)

ESsens_w1= round(mu[0] - (norm.pdf(z_a)/(1-a)) * (Sigma_n@w)[0] * (1/mt.sqrt(w @ Sigma_n @ w.T)),4)
ESsens_w2= round(mu[1] - (norm.pdf(z_a)/(1-a)) * (Sigma_n@w)[1] * (1/mt.sqrt(w @ Sigma_n @ w.T)),4)
ESsens_w3= round(mu[2] - (norm.pdf(z_a)/(1-a)) * (Sigma_n@w)[2] * (1/mt.sqrt(w @ Sigma_n @ w.T)),4)


print(VaRsens_w1)
print(VaRsens_w2)
print(VaRsens_w3)

print('\n')

print(ESsens_w1)
print(ESsens_w2)
print(ESsens_w3)

```
