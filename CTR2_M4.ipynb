```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta, kstest, jarque_bera
```

```python
np.random.seed(1111)
data = beta.rvs(a=0.5,b=0.5, size =500)
```

```python
plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram of Sample Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Perform the Jarque-Bera test

statistic, p_value = jarque_bera(data)
print(f"Jarque-Bera Statistic: {statistic}")
print(f"P-value: {p_value}")

alpha = 0.05 # Significance level

if p_value < alpha:
        print("JB-Test rejects H0: The data is not normally distributed.")
else:
        print("JB-Test fails to reject H0: The data appears to be normally distributed.")
```

```python
statistic2, p_value2 = kstest(data, 'norm')
print(f"Kolmogorov-Smirnov Test: Statistic={statistic2}")
print(f"sp-value={p_value2}")

if p_value2 < alpha:
        print("KS-Test rejects H0: The data is not normally distributed.")
else:
        print("KS-test fails to reject H0: The data appears to be normally distributed.")
```
